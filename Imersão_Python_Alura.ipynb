```python
# prompt: importa a biblioteca pandas

import pandas as pd
import plotly.express as px
```

```python

path_file_excel = "/content/_Imersão Python - Tabela de ações.xlsx"

df_principal= pd.read_excel(path_file_excel, sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel(path_file_excel, sheet_name="Total_de_acoes")
df_total_acoes.head(10)
```

```python
df_ticker = pd.read_excel(path_file_excel, sheet_name="Ticker")
df_ticker.head(10)
```

```python
df_chatgpt = pd.read_excel(path_file_excel, sheet_name="Chatgpt")
df_chatgpt.head(10)
```

```python
df_principal = df_principal[["Ativo","Data","Último (R$)","Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)":"valor_final","Var. Dia (%)":"var_dia_pct"}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal
```

```python
df_principal["valor_inicial"] = df_principal["valor_final"] / (df_principal["Var_pct"]+1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final']-df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
# prompt: aplique essa formula =if(O2>0;"Subiu";if(O2<0;"Desceu";"Estavel")) em uma linha em python

df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker,left_on='Ativo', right_on='Ticker',how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome Empresa', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Nome Empresa'])
df_principal
```

```python
# prompt: você poderia gerar esse código =if(S2>100;"Maior que 100 anos";if(S2<50;"Menor que 50";"Entre 50 e 100 anos")) em python em uma linha

df_principal['Cat_Idade'] = df_principal['Idade (anos)'].apply(lambda x: "Maior que 100 anos" if x > 100 else ("Menor que 50" if x < 50 else "Entre 50 e 100 anos"))
df_principal
```

```python
# prompt: Calcule o maior valor da coluna="Variacao_rs" e imprima o resultado em formato de Reais

maior_variacao_rs = df_principal['Variacao_rs'].max()
print(f'Maior variação em Reais: R${maior_variacao_rs:.2f}')
```

```python
# prompt: Calcule o Menor valor da coluna="Variacao_rs" e imprima o resultado em formato de Reais

menor_variacao_rs = df_principal['Variacao_rs'].min()
print(f'Menor variação em Reais: R${menor_variacao_rs:.2f}')
```

```python
# prompt: Calcule a media valor da coluna="Variacao_rs" e imprima o resultado em formato de Reais

media_variacao_rs = df_principal['Variacao_rs'].mean()
print(f'Média de variação em Reais: R${media_variacao_rs:.2f}')
```

```python
# prompt: Baseado na coluna "Resultado", Calcule os valores da coluna="Variacao_rs"  e imprima a media dos

# Calculate the mean of the "Variacao_rs" column for each unique value in the "Resultado" column
mean_variation_by_result = df_principal.groupby('Resultado')['Variacao_rs'].mean()

# Print the mean variation for each result
for result, mean_variation in mean_variation_by_result.items():
  print(f'Média de variação para {result}: R${mean_variation:.2f}')
```

```python
df_principal_subiu = df_principal[df_principal["Resultado"]=="Subiu"]
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("Segmento")["Variacao_rs"].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo= df_principal.groupby("Resultado")["Variacao_rs"].sum(). reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x="Resultado",y="Variacao_rs", text="Variacao_rs",title="Variação Reais por resultado",)
fig.show()
# Atividade
# Pesquise com a documentação da biblioteca Plotly ou GPT como mudar a formatação dos números do gráfico de barras;
#https://dash.plotly.com/datatable/data-formatting
```

```python
fig_analise_segmento = px.pie(df_analise_segmento,names="Segmento",values="Variacao_rs",title="Variação de quem subiu vs. Segmento")
fig_analise_segmento.show()
```
