```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
dados = yf.download('PETR4.SA', start='2023-01-01', end='2023-12-31')
dados
```

```python
dados.columns
```

```python
dados.columns= ['Abertura', 'Maximo', 'Minimo','Fechamento', 'Fech_Ajust', 'Volume']
```

```python
dados
```

```python
dados = dados.rename_axis('Data')
dados
```

```python
dados['Fechamento'].plot(figsize=(10,6))
plt.title("Variação de preço por data", fontsize=16)
plt.legend(["Fechamento"])
```

```python
df = dados.head(60).copy()
#Converter o indice em uma coluna de data
df['Data'] = df.index

#Converter as datas para o formato numerico de metplotlib
#Isso é necessario para que o metaploitlib possa plotar as datas corretamento nos dados
df['Data'] = df['Data'].apply(mdates.date2num)

df
```

```python
fig, ax = plt.subplots(figsize=(15,8))
#ax == è igual um pincel onde teremos que desenhar coisas na figura que estamos criando

width = 0.7

for i in range(len(df)):
  if df['Fechamento'].iloc[i] > df['Abertura'].iloc[i]:
    color = 'green'
  else:
    color = 'red'

  #Desenhar a linha que fica atras do cadle (Maxima e minima)
  ax.plot([df['Data'].iloc[i], df['Data'].iloc[i]],
          [df['Minimo'].iloc[i], df['Maximo'].iloc[i]],
          color = color,
          linewidth=1)

  #Desenhar o retangulo do cadle por cima da linha que desenhamos na linha acima
  ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width/2, min(df['Abertura'].iloc[i], df['Fechamento'].iloc[i])),
                            width,
                            abs(df['Fechamento'].iloc[i] - df['Abertura'].iloc[i]),
                            facecolor=color))

df['MA7'] = df['Fechamento'].rolling(window=7).mean()
df['MA14'] = df['Fechamento'].rolling(window=14).mean()

#Plotando as médias móveis
ax.plot(df['Data'], df['MA7'], color='blue', label="Média móveis 7 Dias")# Média móveis 7 Dias
ax.plot(df['Data'], df['MA14'], color='yellow', label="Média móveis 14 Dias")# Média móveis 14 Dias

#Adicionar legendas para as médias móveis
ax.legend()


#Formatando o eixo X para mostrar as datas
#Configurando o formato da data e a rotação para melhorar a legibilidade
ax.xaxis_date() # Esse método é usado para dizer ao matploitlib que as datas estão sendo usadas no eixo x
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

#Adicionando titulos e rótulos  para os eixo X e Y
plt.title("Gráfico de Candlestick - PETR4.SA com matplotlib")
plt.xlabel("Data")
plt.ylabel("Preço")

#Adicionando uma grade para falicitar a visulaização dos valores
plt.grid(True)

#Exibindo gráfico
plt.show()

#--------------------------------------------------
#Crinado subplots

fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                    vertical_spacing=0.1,
                    subplot_titles=('Candlestick', 'Volume Transacionado'),
                    row_width=[0.2,0.7])

#Adicionando o gráfico de candlestick
fig.add_trace(go.Candlestick(x=df.index,
                             open=df['Abertura'],
                             high=df['Maximo'],
                             low=df['Minimo'],
                             close=df['Fechamento'],
                             name='Candestick'),
                             row=1, col=1)

#Adicionando as médias Móveis
#Adiconando também as médias móveis ao mesmo subplot para análise de tendências
fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA7'],
                         mode='lines',
                         name='MA7 - Média Móvel 7 Dias'),
                         row=1, col=1)

fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA14'],
                         mode='lines',
                         name='MA14 - Média Móvel 14 Dias'),
                         row=1, col=1)

#Adiconando gráfico da barra para volume
#Em seguida, criamos um gráfico de barras para o volume de transações, que nos da uma idéia de atividade de negociação naquele dia
fig.add_trace(go.Bar(x=df.index,
                     y=df['Volume'],
                     name='Volume'),
                     row=2, col=1)

#Adicionando layout
#Finalmente, configurando  o layout da figura, ajustando títulos, formatos de eixo e outras configurações para tornar o gráfico claro e legendado
fig.update_layout(yaxis_title='Preço',
                  xaxis_rangeslider_visible=False, #Desativar o range slider
                  width=1100, height=600)

fig.show()


```
